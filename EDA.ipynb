```python
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'categorical_column' with the actual column name you want to plot
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

sns.countplot(x='WebsiteTraffic', data=df)
plt.xlabel('Categorical Column')
plt.ylabel('Count')
plt.title('Count of Categorical Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()  # Optional: Ensure that the plot labels are not cut off
plt.show()
```

```python
sns.scatterplot(x='WebsiteTraffic', y='StatsReport', data=df)
plt.xlabel('Numeric Column 1')
plt.ylabel('Numeric Column 2')
plt.title('Scatter Plot of Numeric Column 1 vs. Numeric Column 2')
plt.show()
```

```python
import numpy as np
# Generate random data from an exponential distribution
# You can adjust the parameters (scale and size) as needed
scale_parameter = 1.0  # Adjust as needed
sample_size = 1000  # Adjust as needed
data = np.random.exponential(scale=scale_parameter, size=sample_size)

# Create a histogram to visualize the exponential distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Exponential Distribution')
plt.xlabel('WebsiteTraffic')
plt.ylabel('StatsReport')
plt.title('Histogram of Exponential Distribution')
plt.legend()
plt.grid(True)

# Optionally, you can plot the probability density function (PDF) of the exponential distribution
x = np.linspace(0, max(data), 100)
pdf = (1 / scale_parameter) * np.exp(-x / scale_parameter)
plt.plot(x, pdf, 'r', linewidth=2, label='Exponential PDF')
plt.legend()

plt.show()
```

```python
# Calculate the Pearson correlation matrix for all numeric columns
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(30, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
